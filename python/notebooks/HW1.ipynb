Question 9 

In Colab, create a Pandas dataframe containing four columns and ten rows as
follows:

a) The first column should list ten movies that you have watched.

b) The second column should describe the aspects of the movies that you
liked.

c) The third column should describe the aspects of the movies that you
disliked.

d) The overall rating (a real number between 0 and 5) you’ll allocate to this
movie.

```python
data = {
    'Movie': [
        'The Dark Night', 'Interstellar', 'Moneyball', 
        'Forrest Gump', 'The Big Short', 'Greenland',
        'The Matrix', 'Gladiator', 'Avatar', 'Joker'
        ],
    'Likes': [
        'Amazing Action', 'Really cool visuals', 'Interesting story', 
        'Emotional plot', 'Highlights corruption in banks', 'Intense, action-filled escapes',
        'Innovative sci-fi', 'Epic battles', 'Visual effects', 'Great acting performance'
        ],
    'Dislikes': [
        'Slow pace in the middle', 'Long and slow-moving', 'Historically inaccurate', 
        'Toxic romance', 'Excessive narration and 4th wall breaks', 'Some characters\' choices are frustrating',
        'Dated effects', 'Predictable plot', 'Weak story', 'Dark tone'
        ],
    'Rating': [
        3.9, 5.0, 4.2, 
        4.5, 4.0, 2.3, 
        3.1, 3.0, 2.7, 4.9
        ],   
}

df = pd.DataFrame(data)
print(list(df.columns))
print(df)
```

Question 10 

In a new block, edit the contents of your dataframe from the previous question
as follows:

a) Remove whitespaces from all entries in the “Movie” and “Likes” column,
but not the “Dislikes” column, so “The Dark Knight” becomes “TheDarkKnight”.

b) Make all entries in the “Dislikes” column lowercase, so “Slow pace in the
middle” becomes “slow pace in the middle”.

c) Convert all ratings to their floor values, so a rating of 3.9 becomes 3.

Print the new, edited dataframe.


```python
edited_df = df.copy()

edited_df['Movie'] = df['Movie'].str.replace(' ', '', regex=False)
edited_df['Likes'] = df['Likes'].str.replace(' ', '', regex=False)
edited_df['Dislikes'] = df['Dislikes'].str.lower()
edited_df['Rating'] = np.floor(df['Rating']).astype(int)

print(edited_df)
```

Question 11 

Plot a histogram of the rating column in the original dataframe (using matplotlib). Print the mean and standard deviation of these ratings

```python
plt.hist(df['Rating'], edgecolor='black')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Movie Ratings')
plt.show()

print('Mean:', df['Rating'].mean())
print('Standard Deviation:', df['Rating'].std())
```

Question 12

Plot a 2x1 plot (two rows, one column) of a scatter plot and bar chart visualizing
the ratings from the original dataframe. Hint: use matplotlib.subplots.

```python
fig, axes = plt.subplots(2, 1)

# scatter plot
axes[0].scatter(df['Movie'], df['Rating'])
axes[0].set_title('Scatter Plot of Movie Ratings')
axes[0].set_xlabel('Movie')
axes[0].set_ylabel('Rating')

# bar chart
axes[1].bar(df['Movie'], df['Rating'])
axes[1].set_title('Bar Chart of Movie Ratings')
axes[1].set_xlabel('Movie')
axes[1].set_ylabel('Rating')

plt.tight_layout()
plt.show()
```

Question 13

Create a new column in the DataFrame called “ZRatings” where the original
ratings in the “Rating” column are normalized (Z-scores). Plot a histogram of
the new ratings.

```python
df['ZRating'] = (df['Rating'] - df['Rating'].mean()) / df['Rating'].std()

plt.hist(df['ZRating'], edgecolor='black')
plt.xlabel('ZRatings')
plt.ylabel('Frequency')
plt.title('Normalized Ratings of Movies')
plt.show()
```

Question 14 

Plot a 1x2 plot (one row, two columns) of two histograms, one of the original
ratings (not normalized), and one of the z-scores.


```python
fig, axes = plt.subplots(1,2)

# Histogram of original
axes[0].hist(df['Rating'], edgecolor='black')
axes[0].set_title('Histogram of Original Ratings')
axes[0].set_xlabel('Ratings')
axes[0].set_ylabel('Frequency')

# Histogram of z-scores
axes[1].hist(df['ZRating'], edgecolor='black')
axes[1].set_title('Histogram of Normalized Ratings')
axes[1].set_xlabel('ZRatings')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Question 15 

Add a new column to your dataframe called “RandomRatings”. This column
should contain 10 random real numbers with values between 1 and 100. Repeat
questions 11-14 as follows:

a) Create a histogram of the RandomRatings column, and print the mean
and the standard deviation.

b) Plot a 2x1 plot of a scatter plot and a bar chart visualizing RandomRatings.

c) Create a column called “ZRandomRatings” of z-scores of values in the
“RandomRatings” column and plot a histogram of the Z-values.

d) Plot a 1x2 plot of histograms for the original RandomRatings and ZRandomRatings.

```python
df['RandomRatings'] = np.random.uniform(1, 100, size=len(df))

# a
plt.hist(df['RandomRatings'], edgecolor="black")
plt.xlabel('RandomRatings')
plt.ylabel('Frequency')
plt.title('Histogram of RandomRatings')
plt.show()

print("Mean:", df['RandomRatings'].mean())
print("Standard Deviation:", df['RandomRatings'].std())

# b
fig, axes = plt.subplots(2, 1)

# scatter plot
axes[0].scatter(df['Movie'], df['RandomRatings'])
axes[0].set_title('Scatter Plot of RandomRatings')
axes[0].set_xlabel('Movie')
axes[0].set_ylabel('RandomRatings')

# bar chart
axes[1].bar(df['Movie'], df['RandomRatings'])
axes[1].set_title('Bar Chart of RandomRatings')
axes[1].set_xlabel('Movie')
axes[1].set_ylabel('RandomRatings')

plt.tight_layout()
plt.show()

# c
df['ZRandomRatings'] = (df['RandomRatings'] - df['RandomRatings'].mean()) / df['RandomRatings'].std()

plt.hist(df['ZRandomRatings'], edgecolor='black')
plt.xlabel('ZRandomRatings')
plt.ylabel('Frequency')
plt.title('Normalized RandomRatings of Movies')
plt.show()

# d
fig, axes = plt.subplots(1,2)

# Histogram of original
axes[0].hist(df['RandomRatings'], edgecolor='black')
axes[0].set_title('Histogram of RandomRatings')
axes[0].set_xlabel('RandomRatings')
axes[0].set_ylabel('Frequency')

# Histogram of z-scores
axes[1].hist(df['ZRandomRatings'], edgecolor='black')
axes[1].set_title('Histogram of Normalized RandomRatings')
axes[1].set_xlabel('ZRandomRatings')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
